```python
pValueMax = 0.05
```

```python
x, pValue, x, x = chi2_contingency(automobileData.iloc[:,2::4])
if pValue > pValueMax:
    print('Accept null hypothesis, num of cylinders and acceleration is related')
```

```python
x, pValue, x, x = chi2_contingency(automobileData.iloc[:,1::5])
if pValue > pValueMax:
    print('Accept null hypothesis, mpg and acceleration is related')
```

```python
x, pValue, x, x = chi2_contingency(automobileData.iloc[:,3:5])
if pValue > pValueMax:
    print('Accept null hypothesis, displacement and horsepower is related')
else:
    print('Reject null hypothesis, not related')
```
